# <B><I><U>STANDARDISING CATEGIRICAL DATA </U>----</B></I>

```python
# Check unique values of categorical columns --

categorical_cols = [
    "gender",
    "stress_level",
    "physical_activity",
    "academic_pressure"
]

for col in categorical_cols:
    print(f"\n{col}:")
    print(wellness_df[col].value_counts(dropna=False))
```

```python
# STANDARDISE TEXT VALUES --

for col in categorical_cols:
    wellness_df[col] = (
        wellness_df[col]
        .astype(str)
        .str.strip()
        .str.lower()
    )

print("Categorical text standardized.")
```

```python
for col in categorical_cols:
    print(f"\n{col}:")
    print(wellness_df[col].value_counts(dropna=False))
```

# <B><I><U>ENCODING </U>----</B></I>

```python
# GENDER --

gender_map = {
    "female": 0,
    "male": 1,
    "other": 2
}

wellness_df["gender"] = wellness_df["gender"].map(gender_map)

print("Gender encoded.")
```

```python
# STRESS LEVEL --

stress_map = {
    "low": 0,
    "medium": 1,
    "high": 2
}

wellness_df["stress_level"] = wellness_df["stress_level"].map(stress_map)

print("Stress level encoded.")
```

```python
# PHYSICAL ACTIVITY --

activity_map = {
    "no": 0,
    "yes": 1
}

wellness_df["physical_activity"] = (
    wellness_df["physical_activity"]
    .map(activity_map)
)

print("Physical activity encoded.")
```

```python
# ACADEMIC PRESURE --

pressure_map = {
    "low": 0,
    "medium": 1,
    "high": 2
}

wellness_df["academic_pressure"] = (
    wellness_df["academic_pressure"]
    .map(pressure_map)
)

print("Academic pressure encoded.")
```

# <B><I><U>FEATURE ENGINNERING </U>---</B></I>

```python
# derived features --

# 1. Sleep quality proxy
# ideal sleep near 7.5–8 hours
wellness_df["sleep_balance_score"] = (
    10 - abs(wellness_df["sleep_hours"] - 7.5) * 2
)

wellness_df["sleep_balance_score"] = (
    wellness_df["sleep_balance_score"]
    .clip(lower=0, upper=10)
)

# 2. Digital overload score
wellness_df["digital_overload_score"] = (
    wellness_df["screen_time_hours"] / 12 * 10
).clip(0, 10)

# 3. Study load score
wellness_df["study_load_score"] = (
    wellness_df["study_hours"] / 8 * 10
).clip(0, 10)

# 4. Stress pressure score
wellness_df["stress_pressure_score"] = (
    wellness_df["stress_level"] * 3
    + wellness_df["academic_pressure"] * 2
)

# 5. Lifestyle support score
wellness_df["lifestyle_support_score"] = (
    wellness_df["physical_activity"] * 5
    + wellness_df["sleep_balance_score"] * 0.5
)

print("Derived features created.")
wellness_df.head()
```

```python
# SLEEP DEFICIT --

wellness_df["sleep_deficit"] = (
    8 - wellness_df["sleep_hours"]
).clip(lower=0)
```

```python
#CAFFEIN BURDEN -- 
 
wellness_df["caffeine_burden"] = (
    wellness_df["caffeine_intake"] / 4 * 10
).clip(0, 10)
```

```python
# TARGET () WELLNESS SCORE --

wellness_df["wellness_score"] = (
    50
    + (wellness_df["sleep_balance_score"] * 3.0)
    + (wellness_df["physical_activity"] * 12.0)
    - (wellness_df["stress_level"] * 10.0)
    - (wellness_df["academic_pressure"] * 8.0)
    - (wellness_df["digital_overload_score"] * 1.8)
    - (wellness_df["caffeine_burden"] * 1.2)
    - (wellness_df["study_load_score"] * 1.0)
)

wellness_df["wellness_score"] = (
    wellness_df["wellness_score"]
    .clip(lower=0, upper=100)
    .round(2)
)

wellness_df[[
    "sleep_hours",
    "screen_time_hours",
    "stress_level",
    "physical_activity",
    "academic_pressure",
    "caffeine_intake",
    "wellness_score"
]].head()
```

```python
# SAVING DATASET --

OUTPUT_PATH = PREPROCESSED_DIR / "wellness_feature_engineered.csv"

wellness_df.to_csv(
    OUTPUT_PATH,
    index=False
)

print("Preprocessed wellness dataset saved successfully.")
print("Saved at:", OUTPUT_PATH)
```
