# <b><i><u>student_stress_sleep_screen </u>----</b></i>

<b><i>ALL THE FEATURES OF <U>student_stress_sleep_screen</U> ARE AS FOLLOWS --></b></i>

| Feature             | Type        | Meaning                                     | Wellness me importance                                                                                                                                                                                      |
| ------------------- | ----------- | ------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| `student_id`        | Identifier  | Har student ka unique ID                    | **Model ke liye useful nahi**. Sirf record identify karne ke liye hota hai. Isko model training se **drop** kar denge.                                                                                      |
| `age`               | Numeric     | Student ki age                              | Wellness par thoda indirect effect ho sakta hai, but usually **weak predictor** hota hai jab tak age range bahut wide na ho. Is dataset me 18–25 hi hai, so impact likely low.                              |
| `gender`            | Categorical | Male / Female / Other                       | Lifestyle patterns aur stress response me thoda variation la sakta hai, but usually **secondary feature** hai. Encode karenge aur baad me correlation/importance dekh ke decide karenge rakhna hai ya nahi. |
| `sleep_hours`       | Numeric     | Student kitne ghante sota hai               | **Very important wellness feature**. Achhi sleep = better recovery, mood, focus, lower stress. Wellness score ka major positive driver hoga.                                                                |
| `screen_time_hours` | Numeric     | Daily screen use kitne hours                | **Important negative lifestyle feature**. Zyada screen time often sleep, focus, eye strain, mental fatigue aur routine ko affect karta hai. Wellness ko reduce kar sakta hai.                               |
| `stress_level`      | Categorical | Low / Medium / High stress                  | **Very important**. Stress directly mental wellness ko impact karta hai. Wellness score banate waqt isko negative weight milega.                                                                            |
| `study_hours`       | Numeric     | Daily study hours                           | Moderate study healthy ho sakta hai, but excessive study burnout/stress se linked ho sakta hai. Wellness me **context-dependent** feature hai.                                                              |
| `physical_activity` | Categorical | Student physical activity karta hai ya nahi | **Strong positive wellness feature**. Physical activity mood, sleep, stress reduction aur overall health improve karti hai.                                                                                 |
| `caffeine_intake`   | Numeric     | Daily caffeine amount / cups / intake level | High caffeine sleep aur anxiety ko affect kar sakti hai. **Moderately important negative feature** ho sakta hai, especially stress/sleep ke saath combined.                                                 |
| `academic_pressure` | Categorical | Low / Medium / High academic pressure       | **Important negative feature**. Academic pressure stress aur mental load badhata hai, so wellness reduce kar sakta hai.                                                                                     |
