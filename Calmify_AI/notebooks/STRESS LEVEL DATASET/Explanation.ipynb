# **STRESS LEVEL DATASET** --


<U>--In this Dataset-- FEATURES ARE</U> ->

<U>1. anxiety_level</U> --
 Meaning -
Measures anxiety symptoms.

<B>Higher value = More anxiety.</B>

 <B>Importance</B>-
This will probably be one of the strongest predictor


<U>2. self_esteem</U> --
 Meaning --
Measures confidence and self-worth.

<B>Higher value:</B>
Better self-esteem

<B>Lower value:</B>
Poor self-esteem


<U>3. mental_health_history</U> --
 Meaning--
Binary feature.

Likely:
0 = No previous mental health issues
1 = Previous mental health issues

<B>Importance</B>
People with prior mental health challenges often have higher stress vulnerability


<U>4. depression</U> --
Depression severity score.

<B>Importance --</B>
One of the most influential features.

<B>Often strongly connected to:</B>
stress
burnout
anxiety


<U>5. headache</U> --
Physical symptom score.

<B>Importance</B> --
headaches
migraines
fatigue


<U>6. blood_pressure</U> --
Blood pressure category.

<B>Importance</B> --
High stress can increase blood pressure.

<U>7. sleep_quality</U> --
Sleep quality score.

<B>Usually:</B>
Low sleep quality
      ↓
High stress


<U>8. breathing_problem</U> --
Breathing difficulty severity.

<B>Importance</B>
anxiety attacks
panic episodes


<U>9. noise_level</U> --

Noise in surroundings.

<B>Importance</B>
disturb sleep
affect concentration


<U><B>10. living_conditions</B></U> --
Quality of living environment.

<B>Importance</B>
Poor living conditions increase stress.

<U>11. safety</U> --
Perceived safety.

<B>Importance</B>
Feeling unsafe creates chronic stress.


<U>12. basic_needs</U> --
Food, shelter, resources.

<B>Importance</B>
Very important social determinant.


<U>13. academic_performance</U> --
Academic achievement score.

<B>Importance</B>-
Poor academic performance often increases stress.


<U>14. study_load</U> --
Workload intensity.

<B>Importance</B>-
One of the strongest student stress drivers.


<U>15. teacher_student_relationship</U>--
Relationship quality.

<B>Importance</B>
Supportive teachers reduce stress.


<U>16. future_career_concerns</U> --
Career uncertainty.

<B>Importance</B>
Extremely relevant for college students.

<U>17. social_support</U> --
Support from friends/family.

<B>Importance</B>-
Protective factor.


<U>18. peer_pressure</U>--
Pressure from peers.

<B>Importance</B>-
Can significantly increase stress.


<U>19. extracurricular_activities</U>--
Participation level.

<B>Importance</B>-
Moderate effect.

<B>Can either:</B>
reduce stress
increase stress (if excessive)


<U>20. bullying</U>--
Bullying exposure.

<B>Importance</B>
Very strong mental health predictor.


<U>21. stress_level</U>--
This is what we want to predict.

<B>Likely:</B>

0 = Low
1 = Medium
2 = High

or something similar.
