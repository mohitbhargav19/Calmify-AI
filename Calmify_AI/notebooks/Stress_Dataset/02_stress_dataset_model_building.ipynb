# <b><u><i>stress_dataset_model_building </u>----</b></i>

```python
# ignoring warning --

import warnings
warnings.filterwarnings("ignore")


# importing all necessary libraries --

from pathlib import Path
import pandas as pd
import numpy as np

# preprocessing / model selection
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

try:
    from xgboost import XGBClassifier
    xgb_available = True
except ImportError:
    xgb_available = False
    print("XGBoost not installed. RandomForest and LogisticRegression will still run.")

# evaluation
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix
)

# save model
import joblib

print("warnings ignored:✅✅")
print("ALL NECESSARY LIBRARIES ARE IMPORTED SUCESSFULL ;")
print("🎉🎉🎉🎉")
```

```python

# PATH SETUP


CURRENT_DIR = Path.cwd()                 # notebooks/stress_dataset
NOTEBOOKS_DIR = CURRENT_DIR.parent       # notebooks
PREPROCESSED_DIR = NOTEBOOKS_DIR / "preprocessed"
MODELS_DIR = NOTEBOOKS_DIR / "models"

PREPROCESSED_DIR.mkdir(parents=True, exist_ok=True)
MODELS_DIR.mkdir(parents=True, exist_ok=True)

print("Current Dir      :", CURRENT_DIR)
print("Notebooks Dir    :", NOTEBOOKS_DIR)
print("Preprocessed Dir :", PREPROCESSED_DIR)
print("Models Dir       :", MODELS_DIR)
```

```python
# ============================================================
# LOAD MODEL-READY DATASET
# ============================================================

data_path = PREPROCESSED_DIR / "stress_dataset_model_ready.csv"

df = pd.read_csv(data_path)

print("Dataset loaded successfully")
print("Shape:", df.shape)
df.head()
```

```python
# COLUMNS --

print("Columns in dataset:\n")
for col in df.columns:
    print(col)
```

```python
# TARGET --

TARGET_COL = "stress_type_label"
```

```python
# TARGET DISTRIBUTION --

print("Target column:", TARGET_COL)
print("\nTarget value counts:\n")
print(df[TARGET_COL].value_counts(dropna=False))
```

```python
# FEATURE AND TARGET --

X = df.drop(columns=[TARGET_COL]).copy()
y = df[TARGET_COL].copy()

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# ENCODE TARGET --

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("Encoded classes:")
for i, cls in enumerate(label_encoder.classes_):
    print(i, "->", cls)
```

```python
# TRAIN TEXT SPLIT --

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("Train shape:", X_train.shape, y_train.shape)
print("Test shape :", X_test.shape, y_test.shape)
```

# <B><I><U>MODEL 1</U> ----</B></I>

```python
# LOGISTIC REGRESSION --

log_reg = LogisticRegression(
    max_iter=2000,
    random_state=42
)

log_reg.fit(X_train, y_train)

y_pred_lr = log_reg.predict(X_test)

acc_lr = accuracy_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr, average="weighted")

print("Logistic Regression Accuracy :", round(acc_lr, 4))
print("Logistic Regression Weighted F1 :", round(f1_lr, 4))
```

```python
print("=== Logistic Regression Classification Report ===\n")
print(classification_report(y_test, y_pred_lr))
```

# <B><I><U>MODEL 2</U> ----</B></I>

```python
# RANDOM FOREST --

rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

acc_rf = accuracy_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf, average="weighted")

print("Random Forest Accuracy :", round(acc_rf, 4))
print("Random Forest Weighted F1 :", round(f1_rf, 4))
```

```python
print("=== Random Forest Classification Report ===\n")
print(classification_report(y_test, y_pred_rf))
```

# <B><U><I>MODEL 3</U> ----</B></I>

```python
# XG-BOOST --

if xgb_available:
    xgb_model = XGBClassifier(
        n_estimators=300,
        max_depth=6,
        learning_rate=0.08,
        subsample=0.9,
        colsample_bytree=0.9,
        objective="multi:softmax",
        num_class=len(np.unique(y_encoded)),
        random_state=42,
        eval_metric="mlogloss"
    )

    xgb_model.fit(X_train, y_train)

    y_pred_xgb = xgb_model.predict(X_test)

    acc_xgb = accuracy_score(y_test, y_pred_xgb)
    f1_xgb = f1_score(y_test, y_pred_xgb, average="weighted")

    print("XGBoost Accuracy :", round(acc_xgb, 4))
    print("XGBoost Weighted F1 :", round(f1_xgb, 4))
else:
    acc_xgb = None
    f1_xgb = None
```

```python
if xgb_available:
    print("=== XGBoost Classification Report ===\n")
    print(classification_report(y_test, y_pred_xgb))
```

```python
# MODELS COMPARISON --

results = []

results.append({
    "Model": "Logistic Regression",
    "Accuracy": acc_lr,
    "Weighted_F1": f1_lr
})

results.append({
    "Model": "Random Forest",
    "Accuracy": acc_rf,
    "Weighted_F1": f1_rf
})

if xgb_available:
    results.append({
        "Model": "XGBoost",
        "Accuracy": acc_xgb,
        "Weighted_F1": f1_xgb
    })

results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by=["Accuracy", "Weighted_F1"], ascending=False).reset_index(drop=True)

results_df
```

```python
best_model_name = results_df.loc[0, "Model"]
print("Best model is:", best_model_name)
```

```python
if best_model_name == "Logistic Regression":
    best_model = log_reg
elif best_model_name == "Random Forest":
    best_model = rf
elif best_model_name == "XGBoost":
    best_model = xgb_model
else:
    raise ValueError("Unexpected best model name")
    
print("Selected best model:", best_model_name)
```

```python
# ============================================================
# SAVE BEST MODEL
# ============================================================

model_path = MODELS_DIR / "stress_dataset_best_model.pkl"
encoder_path = MODELS_DIR / "stress_dataset_label_encoder.pkl"
features_path = MODELS_DIR / "stress_dataset_feature_columns.pkl"

joblib.dump(best_model, model_path)
joblib.dump(label_encoder, encoder_path)
joblib.dump(list(X.columns), features_path)

print("Best model saved successfully")
print("Model path   :", model_path)
print("Encoder path :", encoder_path)
print("Features path:", features_path)
```

```python
loaded_model = joblib.load(model_path)
loaded_encoder = joblib.load(encoder_path)
loaded_features = joblib.load(features_path)

print("Loaded model successfully")
print("Loaded encoder classes:", list(loaded_encoder.classes_))
print("Number of features:", len(loaded_features))
```
