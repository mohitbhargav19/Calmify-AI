```python
# =========================
# CLEAN TARGET LABELS
# =========================

df["stress_type"] = df["stress_type"].astype(str).str.strip()

def clean_stress_type(x):
    x_lower = x.lower().strip()

    if "eustress" in x_lower:
        return "eustress"
    elif "distress" in x_lower:
        return "distress"
    elif "no stress" in x_lower:
        return "no_stress"
    else:
        return x_lower

df["stress_type"] = df["stress_type"].apply(clean_stress_type)

print(df["stress_type"].value_counts(dropna=False))
```

```python

# CLEAN TARGET LABELS --

df["stress_type"] = df["stress_type"].astype(str).str.strip()

def clean_stress_type(x):
    x_lower = x.lower()
    
    if "eustress" in x_lower:
        return "eustress"
    elif "distress" in x_lower:
        return "distress"
    else:
        return x_lower

df["stress_type"] = df["stress_type"].apply(clean_stress_type)

print(df["stress_type"].value_counts(dropna=False))
```

```python

# CHECK DUPLICATE ANXIETY COLUMN --


if "anxiety_tension_dup" in df.columns:
    same_values = (df["anxiety_tension"] == df["anxiety_tension_dup"]).all()
    print("Is duplicate anxiety column exactly same?", same_values)
```

```python
if "anxiety_tension_dup" in df.columns:
    if (df["anxiety_tension"] == df["anxiety_tension_dup"]).all():
        df.drop(columns=["anxiety_tension_dup"], inplace=True)
        print("Dropped duplicate column: anxiety_tension_dup")
    else:
        print("Duplicate column not dropped because values differ")
        
print("Current shape:", df.shape)
```

# <b><u><i>FEATURE ENGINEERING </U>----</B></I>

```python

# MENTAL SYMPTOM SCORE --


mental_cols = [
    "recent_stress",
    "anxiety_tension",
    "headaches",
    "irritability",
    "concentration_issues",
    "low_mood",
    "loneliness"
]

df["mental_symptom_score"] = df[mental_cols].mean(axis=1)

df[["mental_symptom_score"]].head()
```

```python

# PHYSICAL STRESS SCORE --

physical_cols = [
    "rapid_heartbeat",
    "sleep_problems",
    "health_issues",
    "weight_change"
]

df["physical_stress_score"] = df[physical_cols].mean(axis=1)

df[["physical_stress_score"]].head()
```

```python
# ACADEMIC STRESS SCORE --


academic_cols = [
    "academic_overload",
    "peer_competition_stress",
    "professor_issues",
    "low_academic_confidence",
    "subject_confidence_issues",
    "academic_extracurricular_conflict"
]

df["academic_stress_score"] = df[academic_cols].mean(axis=1)

df[["academic_stress_score"]].head()
```

```python

# ENVIRONMENT STRES --

environment_cols = [
    "relationship_stress",
    "work_environment_stress",
    "home_environment_stress"
]

df["environment_stress_score"] = df[environment_cols].mean(axis=1)

df[["environment_stress_score"]].head()
```

```python

#  LIFESTYLE STRESS SCORE --


recovery_cols = [
    "sleep_problems",
    "lack_of_relaxation"
]

df["recovery_difficulty_score"] = df[recovery_cols].mean(axis=1)

df[["recovery_difficulty_score"]].head()
```

```python

# OVERALL STRESS RISK SCORE --

overall_components = [
    "mental_symptom_score",
    "physical_stress_score",
    "academic_stress_score",
    "environment_stress_score",
    "recovery_difficulty_score"
]

df["overall_stress_risk_score"] = df[overall_components].mean(axis=1)

df[["overall_stress_risk_score"]].head()
```

```python

# ATTENDANCE RISK --


# Assume higher attendance = better
# So risk = max_scale - attendance
max_attendance_value = df["class_attendance"].max()

df["attendance_risk"] = max_attendance_value - df["class_attendance"]

df[["class_attendance", "attendance_risk"]].head()
```

```python
# COPING DIFFICULTY SCORE

coping_cols = [
    "lack_of_relaxation",
    "loneliness",
    "low_academic_confidence",
    "subject_confidence_issues"
]

df["coping_difficulty_score"] = df[coping_cols].mean(axis=1)

df[["coping_difficulty_score"]].head()
```

```python

# TARGET ENCODING --


target_map = {
    "no_stress": 0,
    "eustress": 1,
    "distress": 2
}

df["stress_type_label"] = df["stress_type"].map(target_map)

print("Encoded target distribution:\n")
print(df["stress_type_label"].value_counts(dropna=False))

print("\nTarget mapping used:")
print(target_map)
```

```python
# =========================
# CHECK UNMAPPED TARGETS
# =========================

unmapped = df[df["stress_type_label"].isnull()]["stress_type"].unique()
print("Unmapped target labels:", unmapped)
```

```python
# =========================
# FINAL MODEL-READY DATAFRAME
# =========================

drop_cols = [
    "stress_type"   # original text target हटाएँगे
]

df_model = df.drop(columns=drop_cols).copy()

print("Model-ready shape:", df_model.shape)
```
