# <B><I><U>FEATURE ENGINEERING </U>----</B></I>

```python
# CREATING HELPING DESK --

positive_words = [
    "happy", "good", "great", "calm", "relaxed", "motivated",
    "productive", "peaceful", "hopeful", "fine", "better", "grateful"
]

negative_words = [
    "sad", "stress", "anxious", "depressed", "angry", "tired",
    "lonely", "worried", "panic", "upset", "cry", "hurt",
    "overwhelmed", "burnout", "hopeless"
]

stress_words = [
    "stress", "pressure", "deadline", "anxiety", "panic",
    "exhausted", "burnout", "tired", "overwhelmed", "worried"
]


def safe_text(x):
    if pd.isna(x):
        return ""
    return str(x).strip().lower()


def count_keywords(text, keywords):
    text = safe_text(text)
    count = 0
    for word in keywords:
        count += text.count(word)
    return count


def word_count(text):
    text = safe_text(text)
    if not text:
        return 0
    return len(text.split())


def char_count(text):
    text = safe_text(text)
    return len(text)
```

```python
# Text-based reflection features --

df["reflection_word_count"] = df["daily_reflections"].apply(word_count)
df["reflection_char_count"] = df["daily_reflections"].apply(char_count)

df["positive_word_count"] = df["daily_reflections"].apply(
    lambda x: count_keywords(x, positive_words)
)

df["negative_word_count"] = df["daily_reflections"].apply(
    lambda x: count_keywords(x, negative_words)
)

df["stress_word_count"] = df["daily_reflections"].apply(
    lambda x: count_keywords(x, stress_words)
)

df[[
    "daily_reflections",
    "reflection_word_count",
    "reflection_char_count",
    "positive_word_count",
    "negative_word_count",
    "stress_word_count"
]].head()
```

```python
#Emotional distress score --

df["emotional_distress_score"] = (
    df["stress_level"] +
    df["anxiety_score"] +
    df["depression_score"]
) / 3
```

```python
# Sleep deficit --


df["sleep_deficit"] = np.maximum(0, 8 - df["sleep_hours"])
```

```python
# Activity score --

df["activity_score"] = df["steps_per_day"] / 1000 
```

```python
# Academic performance risk --

df["gpa_risk"] = 4.0 - df["gpa"]
```

```python
# MOOD INTENSITY --

mood_map = {
    "happy": 2,
    "calm": 2,
    "relaxed": 2,
    "fine": 1,
    "neutral": 1,
    "okay": 1,
    "tired": -1,
    "sad": -2,
    "angry": -2,
    "anxious": -2,
    "stressed": -2,
    "depressed": -3
}

df["mood_description_clean"] = (
    df["mood_description"]
    .astype(str)
    .str.strip()
    .str.lower()
)

df["mood_score"] = df["mood_description_clean"].map(mood_map).fillna(0)
```

```python
# Reflection sentiment imbalance

df["reflection_sentiment_balance"] = (
    df["positive_word_count"] - df["negative_word_count"])
```

```python
# Overall mental risk score --

df["overall_mental_risk_score"] = (
    0.30 * df["stress_level"] +
    0.25 * df["anxiety_score"] +
    0.25 * df["depression_score"] +
    0.10 * df["sleep_deficit"] +
    0.10 * df["gpa_risk"]
)
```

# <B><I><U>ENCODING</U> ----</B></I>

```python
# GENDER --

df["gender"] = df["gender"].astype(str).str.strip().str.lower()

gender_map = {
    "male": 1,
    "female": 0,
    "other": 2
}

df["gender"] = df["gender"].map(gender_map).fillna(2).astype(int)

df["gender"].value_counts()
```

```python
# MOODE DESCRIPTION --

mood_encoder = LabelEncoder()

df["mood_description_encoded"] = mood_encoder.fit_transform(
    df["mood_description_clean"]
)

print("Mood classes:")
for i, cls in enumerate(mood_encoder.classes_):
    print(i, "->", cls)
```
