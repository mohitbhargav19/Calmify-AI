# <b><u><i>mental_health_dataset </u>----</b></i>

<b><i>ALL THE FEATURES OF <U>mental_health_dataset</U> ARE AS FOLLOWS -->

| Feature                  | Type             | Meaning                                                    | Importance for Mental Health Prediction                                                                                                          | Keep / Drop              |
| ------------------------ | ---------------- | ---------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------ | ------------------------ |
| **Student_ID**           | Identifier       | Har student ka unique ID                                   | Model ko koi mental-health pattern nahi deta. Sirf identification ke kaam ka hai.                                                                | **Drop**                 |
| **Age**                  | Numeric          | Student ki age                                             | Different age groups ka stress, sleep, emotional regulation aur academic pressure alag ho sakta hai.                                             | **Keep**                 |
| **Gender**               | Categorical      | Male / Female / Other                                      | Mental health trends aur stress expression gender groups me differ kar sakte hain. Bias handle karte hue useful ho sakta hai.                    | **Keep**                 |
| **GPA**                  | Numeric          | Academic performance / grade point average                 | Low GPA kabhi stress, anxiety, burnout ya low confidence se linked ho sakta hai; high GPA bhi pressure indicate kar sakta hai.                   | **Keep**                 |
| **Stress_Level**         | Numeric          | Student ka overall stress level (e.g. 1–5 scale)           | Mental health prediction ke liye **bahut important** signal hai; high stress often poor mental health status se linked hota hai.                 | **Keep**                 |
| **Anxiety_Score**        | Numeric          | Anxiety symptoms ka score                                  | Mental health condition ka direct predictor hai. Anxiety high ho to status worse hone ke chances zyada hote hain.                                | **Keep**                 |
| **Depression_Score**     | Numeric          | Depression-related symptoms ka score                       | **Very high importance** feature. Mental health status classification me sabse critical features me se ek ho sakta hai.                          | **Keep**                 |
| **Daily_Reflections**    | Text             | Student ka daily journal / reflection text                 | Isme hidden emotional signals hote hain: hopelessness, stress, motivation, loneliness, gratitude, etc. NLP ya sentiment analysis ke liye useful. | **Keep (but transform)** |
| **Sleep_Hours**          | Numeric          | Daily sleep duration                                       | Poor sleep strongly anxiety, low mood, irritability aur poor mental health se linked hoti hai.                                                   | **Keep**                 |
| **Steps_Per_Day**        | Numeric          | Daily physical movement / activity                         | Physical activity mental wellness se positively linked hoti hai; very low steps may indicate low activity / low wellbeing.                       | **Keep**                 |
| **Mood_Description**     | Categorical/Text | Current mood label (Happy, Sad, Tired, etc.)               | Mood directly emotional state represent karta hai. Strong mental health indicator ho sakta hai.                                                  | **Keep**                 |
| **Sentiment_Score**      | Numeric          | Text sentiment score (negative to positive emotional tone) | `Daily_Reflections` ka compressed emotional signal ho sakta hai. Agar reliable hai to very useful feature.                                       | **Keep**                 |
| **Mental_Health_Status** | Target           | Final mental health class/status label                     | Yehi model ka **target variable** hai jise predict karna hai.                                                                                    | **Target**               |
