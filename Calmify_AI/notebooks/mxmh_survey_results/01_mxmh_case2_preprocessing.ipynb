```python
# LOADING DATASET --

# current folder = Calmify_AI/notebooks/mental_health_status
CURRENT_DIR = Path.cwd()

# Calmify_AI root folder
BASE_DIR = CURRENT_DIR.parents[1]

# dataset path = Calmify_AI/notebooks/preprocesssed/mxmh_feature_engineered.csv
DATASET_PATH = BASE_DIR / "datasets" / "mxmh_survey_results.csv"

print("Current Dir :", CURRENT_DIR)
print("Base Dir    :", BASE_DIR)
print("Dataset Path:", DATASET_PATH)

df = pd.read_csv(DATASET_PATH)

print("Dataset loaded successfully:✅✅✅✅")
print("Shape:", df.shape)
df.head()
```

```python
# columns --

df.columns.tolist()
```

```python

# CLEAN COLUMN NAMES --

df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("[", "", regex=False)
      .str.replace("]", "", regex=False)
)

print(df.columns.tolist())
df.head()
```

```python
rename_map = {
    "primary_streaming_service": "streaming_service",
    "hours_per_day": "hours_per_day",
    "while_working": "while_working",
    "foreign_languages": "foreign_languages",
    "fav_genre": "fav_genre",
    "music_effects": "music_effects",

    "frequency_classical": "freq_classical",
    "frequency_country": "freq_country",
    "frequency_edm": "freq_edm",
    "frequency_folk": "freq_folk",
    "frequency_gospel": "freq_gospel",
    "frequency_hip_hop": "freq_hip_hop",
    "frequency_jazz": "freq_jazz",
    "frequency_k_pop": "freq_k_pop",
    "frequency_latin": "freq_latin",
    "frequency_lofi": "freq_lofi",
    "frequency_metal": "freq_metal",
    "frequency_pop": "freq_pop",
    "frequency_r&b": "freq_rnb",
    "frequency_rap": "freq_rap",
    "frequency_rock": "freq_rock",
    "frequency_video_game_music": "freq_video_game_music"
}

df.rename(columns=rename_map, inplace=True)

print(df.columns.tolist())
```

```python

# TARGET CLEANING --

# We only keep valid music effect classes for classification

df["music_effects"] = df["music_effects"].astype(str).str.strip().str.lower()

valid_classes = ["improve", "no effect", "worsen"]

df = df[df["music_effects"].isin(valid_classes)].copy()

print("Shape after target cleaning:", df.shape)
print(df["music_effects"].value_counts(dropna=False))
```

```python
# -----------------------------
# DROP COLUMNS NOT NEEDED
# -----------------------------

drop_cols = ["permissions"]

# timestamp ko later features ke liye use nahi karna hai case 2 me
# so directly drop kar sakte hain
if "timestamp" in df.columns:
    drop_cols.append("timestamp")

existing_drop_cols = [col for col in drop_cols if col in df.columns]
df.drop(columns=existing_drop_cols, inplace=True)

print("Dropped columns:", existing_drop_cols)
print("Current shape:", df.shape)
```

```python

# MISSING VALUES --

missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]

print("Missing values:\n")
print(missing)
```

```python

# HANDLE MISSING VALUES --

# numeric columns
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

# categorical columns
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()

# fill numeric with median
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# fill categorical with mode
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values after filling:")
print(df.isnull().sum().sum())
```

```python
# -----------------------------
# CLEAN TEXT VALUES
# -----------------------------

for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()

```

# <b><u><i>FEATURE ENGINEERED </U>----</B></I>

```python
# -----------------------------
# MAP MUSIC FREQUENCY COLUMNS
# -----------------------------

frequency_map = {
    "never": 0,
    "rarely": 1,
    "sometimes": 2,
    "very frequently": 3
}

freq_cols = [
    "freq_classical", "freq_country", "freq_edm", "freq_folk",
    "freq_gospel", "freq_hip_hop", "freq_jazz", "freq_k_pop",
    "freq_latin", "freq_lofi", "freq_metal", "freq_pop",
    "freq_rnb", "freq_rap", "freq_rock", "freq_video_game_music"
]

for col in freq_cols:
    df[col] = df[col].map(frequency_map)

print(df[freq_cols].head())
```

```python

# MAP YES/NO COLUMNS --

yes_no_map = {
    "yes": 1,
    "no": 0
}

binary_cols = [
    "while_working",
    "instrumentalist",
    "composer",
    "exploratory",
    "foreign_languages"
]

for col in binary_cols:
    df[col] = df[col].map(yes_no_map)

print(df[binary_cols].head())
```

```python

# COMPOSITE MENTAL SYMPTOM FEATURE --

df["mental_symptom_score"] = (
    df["anxiety"] +
    df["depression"] +
    df["insomnia"] +
    df["ocd"]
)

df["mental_symptom_avg"] = df["mental_symptom_score"] / 4

df[["anxiety", "depression", "insomnia", "ocd",
    "mental_symptom_score", "mental_symptom_avg"]].head()
```

```python
# -----------------------------
# MUSIC ENGAGEMENT SCORE
# -----------------------------

df["music_engagement_score"] = (
    df["hours_per_day"] +
    df["while_working"] +
    df["instrumentalist"] +
    df["composer"] +
    df["exploratory"]
)

df[["hours_per_day", "while_working", "instrumentalist",
    "composer", "exploratory", "music_engagement_score"]].head()
```

```python

# GENRE DIVERSITY SCORE --

# kitne genres user at least sometimes sunta hai

df["genre_diversity_score"] = (df[freq_cols] >= 2).sum(axis=1)

df[["genre_diversity_score"]].head()
```

```python

# CALM MUSIC SCORE --

# sleep / anxiety support ke liye useful genres ka aggregate

calm_cols = ["freq_classical", "freq_lofi", "freq_jazz", "freq_folk"]

df["calm_music_score"] = df[calm_cols].mean(axis=1)

df[["calm_music_score"]].head()
```

```python

# HIGH ENERGY MUSIC SCORE --

energy_cols = ["freq_edm", "freq_hip_hop", "freq_rap", "freq_metal", "freq_rock"]

df["high_energy_score"] = df[energy_cols].mean(axis=1)

df[["high_energy_score"]].head()
```

```python

mainstream_cols = ["freq_pop", "freq_rnb", "freq_latin", "freq_country"]

df["mainstream_music_score"] = df[mainstream_cols].mean(axis=1)

df[["mainstream_music_score"]].head()
```

```python

# BPM CLEANING --


df["bpm"] = df["bpm"].clip(lower=40, upper=250)

df["bpm"].describe()
```

```python

# BPM BAND --

df["bpm_band"] = pd.cut(
    df["bpm"],
    bins=[0, 90, 120, 160, 300],
    labels=["low_bpm", "medium_bpm", "high_bpm", "very_high_bpm"]
)

df["bpm_band"].value_counts(dropna=False)
```

```python

# LISTENING BAND --

df["listening_band"] = pd.cut(
    df["hours_per_day"],
    bins=[-1, 2, 5, 24],
    labels=["low_listener", "moderate_listener", "heavy_listener"]
)

df["listening_band"].value_counts(dropna=False)
```

```python

# MENTAL HEALTH BAND --

df["mental_health_band"] = pd.cut(
    df["mental_symptom_score"],
    bins=[-1, 10, 20, 40],
    labels=["low", "moderate", "high"]
)

df["mental_health_band"].value_counts(dropna=False)
```

```python

# TARGET ENCODING --

target_map = {
    "improve": 0,
    "no effect": 1,
    "worsen": 2
}

df["music_effects_label"] = df["music_effects"].map(target_map)

print(df[["music_effects", "music_effects_label"]].head())
print(df["music_effects_label"].value_counts())
```

```python


raw_symptom_cols = ["anxiety", "depression", "insomnia", "ocd"]

df.drop(columns=raw_symptom_cols, inplace=True)

print("Dropped raw symptom columns:", raw_symptom_cols)
print("Current shape:", df.shape)
```

```python


categorical_cols_after_fe = df.select_dtypes(include=["object", "category"]).columns.tolist()

print("Categorical columns before encoding:")
print(categorical_cols_after_fe)
```

```python
df = pd.get_dummies(
    df,
    columns=categorical_cols_after_fe,
    drop_first=False
)

print("Shape after encoding:", df.shape)
df.head()
```

```python

# REMOVE DUPLICATE COLUMNS --

df = df.loc[:, ~df.columns.duplicated()].copy()

print("Shape after removing duplicate columns:", df.shape)
```
