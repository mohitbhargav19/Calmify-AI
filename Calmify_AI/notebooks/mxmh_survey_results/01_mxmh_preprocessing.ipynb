```python

# CLEAN COLUMN NAMES --

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("[", "", regex=False)
    .str.replace("]", "", regex=False)
    .str.replace("/", "_")
)

print(df.columns.tolist())
```

```python

# RENAME COLUMNS --

rename_map = {
    "primary_streaming_service": "streaming_service",
    "hours_per_day": "hours_per_day",
    "while_working": "while_working",
    "fav_genre": "fav_genre",
    "foreign_languages": "foreign_languages",
    "music_effects": "music_effects",

    "frequency_classical": "freq_classical",
    "frequency_country": "freq_country",
    "frequency_edm": "freq_edm",
    "frequency_folk": "freq_folk",
    "frequency_gospel": "freq_gospel",
    "frequency_hip_hop": "freq_hip_hop",
    "frequency_jazz": "freq_jazz",
    "frequency_k_pop": "freq_k_pop",
    "frequency_latin": "freq_latin",
    "frequency_lofi": "freq_lofi",
    "frequency_metal": "freq_metal",
    "frequency_pop": "freq_pop",
    "frequency_r&b": "freq_rnb",
    "frequency_rap": "freq_rap",
    "frequency_rock": "freq_rock",
    "frequency_video_game_music": "freq_vgm"
}

df.rename(columns=rename_map, inplace=True)

print(df.columns.tolist())
```

```python

# DROP CLEARLY UNNECESSARY COLUMNS --

drop_cols = []

if "timestamp" in df.columns:
    drop_cols.append("timestamp")

if "permissions" in df.columns:
    drop_cols.append("permissions")

df.drop(columns=drop_cols, inplace=True)

print("Dropped columns:", drop_cols)
print("New shape:", df.shape)
```

```python

# DUPLICATE CHECK --

duplicate_count = df.duplicated().sum()
print("Duplicate rows:", duplicate_count)

if duplicate_count > 0:
    df = df.drop_duplicates().reset_index(drop=True)
    print("Duplicates removed.")
    print("New shape:", df.shape)
```

```python

# MISSING VALUES --

missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]

print("Missing values after initial cleanup:")
print(missing)
```

```python

# HANDLE MISSING VALUES --

numeric_fill_median = ["age", "hours_per_day", "bpm"]

for col in numeric_fill_median:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

categorical_cols = df.select_dtypes(include="object").columns.tolist()

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values after imputation:")
print(df.isnull().sum().sum())
```

```python

# CHECK UNIQUE VALUES --

check_cols = [
    "streaming_service", "while_working", "instrumentalist", "composer",
    "fav_genre", "exploratory", "foreign_languages", "music_effects"
]

for col in check_cols:
    if col in df.columns:
        print(f"\n{col}")
        print(df[col].value_counts(dropna=False))
```

```python

# CLEAN STRING VALUES --

obj_cols = df.select_dtypes(include="object").columns.tolist()

for col in obj_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()

print("String cleaning done.")
```

```python

# CLEAN BPM --

if "bpm" in df.columns:
    bpm_median = df.loc[(df["bpm"] > 0) & (df["bpm"] <= 250), "bpm"].median()

    df.loc[df["bpm"] <= 0, "bpm"] = np.nan
    df.loc[df["bpm"] > 250, "bpm"] = np.nan

    df["bpm"] = df["bpm"].fillna(bpm_median)

print(df["bpm"].describe())
```

```python

# YES / NO ENCODING --

yes_no_cols = [
    "while_working",
    "instrumentalist",
    "composer",
    "exploratory",
    "foreign_languages"
]

yes_no_map = {
    "yes": 1,
    "no": 0
}

for col in yes_no_cols:
    if col in df.columns:
        df[col] = df[col].map(yes_no_map)

print(df[yes_no_cols].head())
```

```python

# ENCODE FREQUENCY COLUMNS --

frequency_cols = [
    "freq_classical", "freq_country", "freq_edm", "freq_folk", "freq_gospel",
    "freq_hip_hop", "freq_jazz", "freq_k_pop", "freq_latin", "freq_lofi",
    "freq_metal", "freq_pop", "freq_rnb", "freq_rap", "freq_rock", "freq_vgm"
]

freq_map = {
    "never": 0,
    "rarely": 1,
    "sometimes": 2,
    "very frequently": 3
}

for col in frequency_cols:
    if col in df.columns:
        df[col] = df[col].map(freq_map)

print(df[frequency_cols].head())
```

```python

# ENCODE MUSIC EFFECTS TARGET --

music_effect_map = {
    "improve": 2,
    "no effect": 1,
    "worsen": 0
}

if "music_effects" in df.columns:
    df["music_effects_label"] = df["music_effects"].map(music_effect_map)

print(df[["music_effects", "music_effects_label"]].head())
print(df["music_effects_label"].value_counts(dropna=False))
```

```python
# ONE-HOT ENCODING --

nominal_cols = ["streaming_service", "fav_genre"]

for col in nominal_cols:
    if col in df.columns:
        print(f"\n{col} unique values:")
        print(sorted(df[col].dropna().unique().tolist()))
```

# <b><u><i> FEATURE ENGINEERING</U> ----</B></I>

```python

# 1. mental health score
df["mental_health_score"] = (
    df["anxiety"] +
    df["depression"] +
    df["insomnia"] +
    df["ocd"]
) / 4

# 2. music engagement score
df["music_engagement_score"] = (
    df["hours_per_day"] +
    df["exploratory"] +
    df["instrumentalist"] +
    df["composer"]
)

# 3. focus listening score
df["focus_listening_score"] = (
    df["while_working"] +
    df["freq_lofi"] +
    df["freq_classical"] +
    df["freq_vgm"]
)

# 4. high energy listening score
df["high_energy_score"] = (
    df["freq_edm"] +
    df["freq_rap"] +
    df["freq_hip_hop"] +
    df["freq_pop"]
)

# 5. calm music score
df["calm_music_score"] = (
    df["freq_classical"] +
    df["freq_lofi"] +
    df["freq_jazz"] +
    df["freq_folk"]
)

# 6. intense music score
df["intense_music_score"] = (
    df["freq_metal"] +
    df["freq_rock"] +
    df["freq_rap"]
)

# 7. total genre diversity score
df["genre_diversity_score"] = df[frequency_cols].sum(axis=1)

# 8. bpm_energy_zone
def bpm_zone(x):
    if x < 90:
        return "low_bpm"
    elif x <= 130:
        return "medium_bpm"
    else:
        return "high_bpm"

df["bpm_zone"] = df["bpm"].apply(bpm_zone)

print("Feature engineering completed.")
```

```python

# MENTAL HEALTH SEVERITY BAND --

def mental_band(score):
    if score < 3:
        return "low"
    elif score < 6:
        return "moderate"
    else:
        return "high"

df["mental_health_band"] = df["mental_health_score"].apply(mental_band)

print(df["mental_health_band"].value_counts())
```

```python

# LISTENING INTENSITY BAND --

def listening_band(hours):
    if hours < 2:
        return "low_listener"
    elif hours < 5:
        return "moderate_listener"
    else:
        return "heavy_listener"

df["listening_band"] = df["hours_per_day"].apply(listening_band)

print(df["listening_band"].value_counts())
```

```python

# PREPARE MODEL-READY DATASET --

df_model = df.copy()

# columns to drop from model data
drop_model_cols = [
    "music_effects"   # original text target
]

df_model.drop(columns=[c for c in drop_model_cols if c in df_model.columns], inplace=True)

# one-hot encode nominal categorical columns
one_hot_cols = [
    "streaming_service",
    "fav_genre",
    "bpm_zone",
    "mental_health_band",
    "listening_band"
]

df_model = pd.get_dummies(
    df_model,
    columns=[c for c in one_hot_cols if c in df_model.columns],
    drop_first=True
)

print("Model-ready shape:", df_model.shape)
df_model.head()
```
