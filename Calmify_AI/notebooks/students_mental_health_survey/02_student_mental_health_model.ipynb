```python
# IMPORTING ALL THE NECESSARY LIBRARIES --

# Data Handling

import pandas as pd
import numpy as np

# Visualization

import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing

from sklearn.preprocessing import (
    LabelEncoder,
    StandardScaler,
    MinMaxScaler
)

# Train Test Split

from sklearn.model_selection import (
    train_test_split,
    GridSearchCV,
    cross_val_score
)

# Models

from sklearn.ensemble import (
    RandomForestClassifier,
    RandomForestRegressor
)

from xgboost import (
    XGBClassifier,
    XGBRegressor
)

# Evaluation

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# Model Saving

import joblib

# Display

pd.set_option(
    "display.max_columns",
    None
)

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier

sns.set_style("whitegrid")

print("all the necessary libraries are imported sucessfully:")
print("🎉🎉🎉🎉")
```

```python
# PATH SETUP
# current notebook folder
CURRENT_DIR = Path.cwd()

# notebooks folder
NOTEBOOKS_DIR = CURRENT_DIR.parent

# project root
BASE_DIR = NOTEBOOKS_DIR.parent

# folders
PREPROCESSED_DIR = NOTEBOOKS_DIR / "preprocessed"
MODELS_DIR = NOTEBOOKS_DIR / "models"

PREPROCESSED_DIR.mkdir(parents=True, exist_ok=True)
MODELS_DIR.mkdir(parents=True, exist_ok=True)

# file path
DATA_PATH = PREPROCESSED_DIR / "students_mental_health_survey_model_ready.csv"

print("Current Dir:", CURRENT_DIR)
print("Notebooks Dir:", NOTEBOOKS_DIR)
print("Preprocessed Dir:", PREPROCESSED_DIR)
print("Models Dir:", MODELS_DIR)
print("Data Path:", DATA_PATH)
```

```python
# LOADING DATASET --

df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
df.head()
```

```python
# BASIC INSPECTION --

print("Columns:")
print(df.columns.tolist())

print("\nInfo:")
print(df.info())

print("\nTarget distribution:")
print(df["stress_level"].value_counts().sort_index())
```

```python
# DEFINING X AND Y --

TARGET_COL = "stress_level"

X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# FEATURE NAME --

feature_names = X.columns.tolist()

print("Total features:", len(feature_names))
feature_names[:20]
```

```python
# TRAIN TEST SPLIT --

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

# <B><U><I>MODEL 1 </U>----

```python
# LOGGISTIC REGRESSION --

log_reg_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        max_iter=3000,
        random_state=42
    ))
])
```

```python
log_reg_pipeline.fit(X_train, y_train)
```

```python
# EVALUATION --

y_pred_lr = log_reg_pipeline.predict(X_test)

lr_acc = accuracy_score(y_test, y_pred_lr)
lr_f1 = f1_score(y_test, y_pred_lr, average="weighted")

print("Logistic Regression Accuracy:", round(lr_acc * 100, 2), "%")
print("Logistic Regression Weighted F1:", round(lr_f1, 4))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_lr))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, y_pred_lr))
```

# <B><I><U>MODEL 2</U> ----</B></I>

```python
# RANDOM FOREST --

rf_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)
```

```python
rf_model.fit(X_train, y_train)
```

```python
# EVALUATION --

y_pred_rf = rf_model.predict(X_test)

rf_acc = accuracy_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf, average="weighted")

print("Random Forest Accuracy:", round(rf_acc * 100, 2), "%")
print("Random Forest Weighted F1:", round(rf_f1, 4))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_rf))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, y_pred_rf))
```

# <B><U><I>MODEL 3</U> ----</B></I>

```python
# XG_BOOST --

xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.9,
    colsample_bytree=0.9,
    objective="multi:softmax",
    num_class=len(np.unique(y)),
    random_state=42,
    eval_metric="mlogloss"
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
# EVALUATION --

y_pred_xgb = xgb_model.predict(X_test)

xgb_acc = accuracy_score(y_test, y_pred_xgb)
xgb_f1 = f1_score(y_test, y_pred_xgb, average="weighted")

print("XGBoost Accuracy:", round(xgb_acc * 100, 2), "%")
print("XGBoost Weighted F1:", round(xgb_f1, 4))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_xgb))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, y_pred_xgb))
```

```python
# MODEL COMPARISON --

results = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "XGBoost"
    ],
    "Accuracy": [
        lr_acc,
        rf_acc,
        xgb_acc
    ],
    "Weighted_F1": [
        lr_f1,
        rf_f1,
        xgb_f1
    ]
})

results = results.sort_values(by=["Accuracy", "Weighted_F1"], ascending=False).reset_index(drop=True)
results
```
