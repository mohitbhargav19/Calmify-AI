```python
# standardising columns --

df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

print(df.columns.tolist())
```

```python
# DUPLICATES --

duplicate_count = df.duplicated().sum()
print("Duplicate rows:", duplicate_count)
```

```python
# MISSING VALUES --

missing = df.isnull().sum().sort_values(ascending=False)
missing[missing > 0]
```

```python
# HANDLING MISSING VALUES --

df["cgpa"] = df["cgpa"].fillna(df["cgpa"].median())

df["substance_use"] = df["substance_use"].fillna(df["substance_use"].mode()[0])
```

```python
df.isnull().sum()
```

# <B><I><U>ENCODING </U> ----</B></I>

```python
# CATEGORICAL COLUMNS --

categorical_cols = df.select_dtypes(include="object").columns.tolist()

for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts(dropna=False))
```

```python
# CLEAN TEXT COLUMNS --

for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
# ORDINAL MAPPING --

# quality / level mappings
quality_map = {
    "Poor": 0,
    "Average": 1,
    "Good": 2
}

activity_map = {
    "Low": 0,
    "Moderate": 1,
    "High": 2
}

support_map = {
    "Low": 0,
    "Moderate": 1,
    "High": 2
}

substance_map = {
    "Never": 0,
    "Occasionally": 1,
    "Frequently": 2
}
```

```python
# ORDINAL ENCODING --

df["sleep_quality"] = df["sleep_quality"].map(quality_map)
df["physical_activity"] = df["physical_activity"].map(activity_map)
df["diet_quality"] = df["diet_quality"].map(quality_map)
df["social_support"] = df["social_support"].map(support_map)
df["substance_use"] = df["substance_use"].map(substance_map)
df["counseling_service_use"] = df["counseling_service_use"].map(substance_map)
df["extracurricular_involvement"] = df["extracurricular_involvement"].map(activity_map)
```

```python
# BINARY COLUMNS --

binary_map = {
    "No": 0,
    "Yes": 1
}

df["family_history"] = df["family_history"].map(binary_map)
df["chronic_illness"] = df["chronic_illness"].map(binary_map)
```

```python
# GENDER --

gender_map = {
    "Male": 1,
    "Female": 0
}

df["gender"] = df["gender"].map(gender_map)

# unknown / others -> 2
df["gender"] = df["gender"].fillna(2).astype(int)
```

```python
# RELATIONSHIP STATUS --

relationship_map = {
    "Single": 0,
    "In a Relationship": 1,
    "Married": 2
}

df["relationship_status"] = df["relationship_status"].map(relationship_map)
```

```python
# RESIDENCE TYPE--

# Check unique values first
print(df["residence_type"].value_counts(dropna=False))

residence_map = {
    "On-Campus": 0,
    "Off-Campus": 1,
    "With Family": 2,
    "Hostel": 0,          # fallback if present
    "Home": 2             # fallback if present
}

df["residence_type"] = df["residence_type"].map(residence_map)
```

```python
# COURSE --

course_dummies = pd.get_dummies(
    df["course"],
    prefix="course",
    drop_first=True
)

df = pd.concat([df, course_dummies], axis=1)
df.drop(columns=["course"], inplace=True)

print("Shape after course encoding:", df.shape)
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

# <B><I><U>FEATURE ENGINEERING</U> ----</B></I>

```python
# Mental health combined score --

df["mental_health_score"] = (
    df["depression_score"] + df["anxiety_score"]
)
```

```python
# WELLNESS SCORE --

df["wellness_score"] = (
    df["sleep_quality"] +
    df["physical_activity"] +
    df["diet_quality"] +
    df["social_support"] +
    df["extracurricular_involvement"]
)
```

```python
# ACADEMIC LOAD --

# cgpa ko reverse stress signal ke saath use karenge:
# lower cgpa -> more academic concern
df["cgpa_risk"] = 4.0 - df["cgpa"]

df["academic_load_score"] = (
    df["cgpa_risk"] +
    df["semester_credit_load"] / 10 +
    df["financial_stress"]
)
```

```python
# HEALTH RISK --

df["health_risk_score"] = (
    df["family_history"] +
    df["chronic_illness"] +
    df["substance_use"] +
    df["counseling_service_use"]
)
```

```python
# SOCIAL RISK --

# lower support / lower involvement / difficult relationship can add stress
df["social_risk_score"] = (
    (2 - df["social_support"]) +
    (2 - df["extracurricular_involvement"]) +
    df["relationship_status"]
)
```

```python
# OVERALL RISK --

df["overall_risk_score"] = (
    df["mental_health_score"] +
    df["academic_load_score"] +
    df["health_risk_score"] +
    df["social_risk_score"]
)
```

```python
# STRESS VULNARABILITY --

df["stress_vulnerability_score"] = (
    df["depression_score"] +
    df["anxiety_score"] +
    df["financial_stress"] +
    (2 - df["sleep_quality"])
)
```
